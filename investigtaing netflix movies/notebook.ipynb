| `duration` | Duration of the show in minutes |
| `description` | Description of the show |
| `genre` | Show genre |

```python
# Importing pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# checking the data
netflix_df.info()
```

```python
# filtering the release dates to 1990s
df90s = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] <= 1999)]
```

```python
# visualising the data to get the most frequent movie in 1990's
plt.hist(df90s['duration'], bins = 50)
plt.xlabel("duration of the movie")
plt.ylabel("number of movies")
plt.xticks(np.arange(0, 210, step = 10))
plt.show()
# the most frequent duration is approximately equals 95
```

```python
#getting the most frequent duration using mode
duration = stats.mode(df90s['duration']).mode[0]
print(duration)
```

```python
# counting movies who's duration is less than 90 min
short_movie_count = ((df90s['duration'] < 90) & (df90s['genre'] == 'Action')).sum()
print(short_movie_count)
```
